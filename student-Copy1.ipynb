### Review Normality ###

```python
# Check against normal distribution - both skew right
# P-value = 0.0 meaning that it is impossible for the sample to be part of 
#a normal distribution?
def normal_Xy(X, y):
    fig, axes = plt.subplots(ncols=2, figsize=(10,5))
                             
    sns.distplot(X, ax=axes[0], kde=False, bins = 'auto')
    axes[0].set(xlabel= 'X')
    sns.distplot(y, ax=axes[1], kde=False, bins='auto', color='r')
    axes[1].set(xlabel= 'y')
    print("X Normality: ",stats.normaltest(X))
    print("y Normalty: ",stats.normaltest(y))
    return fig, axes
normal_Xy(X, y)
```

#### Both are skewed right...again outliers? ####
