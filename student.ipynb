```python
# formula based regression
import statsmodels.formula.api as smf
df = df_clean.copy()
df.head(2)
```

```python
def scatter_matrix(data):
    pd.plotting.scatter_matrix(data);
scatter_matrix(df)
```

```python
df.head()
```

```python
# Create Model Function
def model_in(df, target, removed = []):

# Concat column names
#    df.drop([target], axis =1, inplace=True)
    f = list(df.columns)
    for col in removed:
        if col in f:
            f.remove(col)
    f = '+'.join(f)
    display(f)
#Create formula
    formula= target + '~' + f
    display(formula)
    model = smf.ols(formula, df).fit()
    display(model.summary())
    fig = sm.graphics.qqplot(model.resid, dist=stats.norm,fit=True, line='45')
    return model
    return model.summary
    return model.params()
    
    
model_in(df, 'price_est', removed=['price_est'])
```
