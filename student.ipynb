```python
#create function to calculate z-scores and keep those > 3 which are the 
# statistical outliers


def outliers_z(data):

    """ Computes the outliers for all features(columns) in a dataframe and
        uses a filter to return only items in the columns with a z_score > 3.
​
    Parameters
    ----------------------------------------
    data[df]::
        The dataframe being analyzed
    
​
    Returns
    ----------------------------------------
    all outliers with a z_score > 3 for all columns in a dataframe
​
    Example Usage
    ----------------------------------------
    >>> global_outliers #to use the outliers results outside of
    >>>   the dataframe
    >>> z_calc = pd.DataFrame(z_calc, index= data.index, columns = 
    >>>   data.columns)
    >>> outliers = pd.DataFrame(z_calc > 3, index = data.index, columns =
    >>>   data.columns)
    >>> outliers_df = outliers.any(axis=1)
    >>> return outliers_df
    """

    global outliers_df
    z_calc = np.abs(stats.zscore(data))
    z_calc = pd.DataFrame(z_calc, index = data.index, columns=data.columns)
    outliers = pd.DataFrame(z_calc>3, index = data.index, columns=data.columns) 
    outliers_df = outliers.any(axis=1)
    return outliers_df

outliers_z(df)
outliers_df.value_counts()
```

```python
# Check against normal distribution - both skew right

""" Computes the normality of all columns given(usually the entire data-
        frame)
​
    Parameters
    ----------------------------------------
    df: dataframe used for analysis
    columns: usually all columns in the dataframe

​
    Returns
    ----------------------------------------
    returns plots of normality for all columns in a dataframe.  Heading of 
    plot show normality % and distplot shows the normality of the data in
    each column
​
    Example Usage
    ----------------------------------------
    >>> for item in df.columns::
    >>>     X = df[item]
    >>>     fig. ax = plt.subplits(figsize = (8, 5))
    >>>     sns.distplot(X, ax=ax, kde=False, bins='auto')
    >>>     ax.set(xlabel = item)
    >>>     stat, p = statsnormal test(X)
    >>>     ax.set_title(f"{item} Normality: {round(3)}")
    """


# P-value = 0.0 meaning that features and target are not normal
def normal_Xy(df, columns):
    for item in columns:
        X = df[item]
        fig, ax = plt.subplots(figsize=(8,5))                         
        sns.distplot(X, ax=ax, kde=False, bins = 'auto')
        ax.set(xlabel= item)
        stat, p =  stats.normaltest(X)
        ax.set_title(f"{item} Normality: {round(p, 3)}" )
    return fig, ax
normal_Xy(df, df.columns)
```

```python
# with help from James Irving

def model_in(df, target, removed = []):

    """ Computes and shows the results of the OLS regression model
​
    Parameters
    ----------------------------------------
    data[df, target::
        The dataframe to be analyzed
    target:: 
        The dependent or y variable
    removed = []::
        input the target to be removed
​
    Returns
    ----------------------------------------
    inputs and results of OLS Regression Model.  Inputs = calcualtion and
    display of the concatonated columns of the features(columns) of a
    dataframe to be used in the formula of the OLS model.  Outputs = 
    calculation and return of the OLS Model results, Model Summary and a
    qqplot
​
    Example Usage
    ----------------------------------------
    >>> # concat column names
    >>> f = list(df.columns)
    >>> for col in removed::
    >>>     if col in f::
    >>>         f.remove(col)
    >>>     f = '+'.join(f)
    >>> # create formula
    >>> formula = target + '-' + f
    >>> display(formula)
    >>> moodel = smf.ols(formula, df).fit()
    >>> display(model.summary())
    >>> # create qq plot
    >>> fig = sm.graphics.qqplot(model, resid, dist = stats.norm, fit =
    >>>   True, line='45')
    >>> return model
    >>> return model.summary
    """

# Concat column names
#    df.drop([target], axis =1, inplace=True)
    f = list(df.columns)
    for col in removed:
        if col in f:
            f.remove(col)
    f = '+'.join(f)
    display(f)
#Create formula
    formula= target + '~' + f
    display(formula)
    model = smf.ols(formula, df).fit()
    display(model.summary())
    fig = sm.graphics.qqplot(model.resid, dist=stats.norm,fit=True, line='45')
    
    return model
    return model.summary
    
    
    
model_in(df, 'price_est', removed=['price_est'])
```
